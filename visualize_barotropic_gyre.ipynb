```python
# check volume conservations
U_bt = ds.U * ds.drF * ds.hFacW
V_bt = ds.V * ds.drF * ds.hFacS

dU = grid.diff(U_bt, 'X', boundary='fill')
dV = grid.diff(V_bt, 'Y', boundary='fill')
divergence = (dU + dV) / ds.rA

fig, axs = plt.subplots(figsize=(12,4), ncols=2)
dU[-1].plot(ax=axs[0])
dV[-1].plot(ax=axs[1])


plt.figure()
divergence[-1].plot()
```

```python
beta = 1e-11
r = 1e-3
Ah = 100.
rho0 = 1000.
```

```python
def curl(u, v):
    return (-grid.diff(u * ds.dxC * ds.hFacW, 'Y', boundary='extend') +
             grid.diff(v * ds.dyC * ds.hFacS, 'X', boundary='extend')) / ds.rAz
```

```python
Z = curl(U_bt, V_bt)
bot_drag = -r * Z / ds.drF
bot_drag[-1].plot()
```

```python
# vorticity budget
beta_V = beta * V_bt
beta_V[-1].plot()
```
